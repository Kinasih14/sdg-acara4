```python
##mengurutkan data dari besar ke kecil
sorted_country_df = country_df.sort_values('confirmed',
ascending=False).head(25)
```

```python
sorted_country_df ##mengurutkan data nama negara dari terbesar ke terkecil
```

```python
##membuat style untuk tabel (memberi warna pada kolom)
def highlight_col(x):
    o = 'background-color: orange'
    r = 'background-color: red'
    g = 'background-color: green'
    temp_df= pd.DataFrame('', index=x.index, columns = x.columns) 
    temp_df.iloc[:,4] = o
    temp_df.iloc[:,5] = r
    temp_df.iloc[:,6] = g 
    return temp_df
```

```python
##melakukan pengurutan negara berdasarkan data terkonformasi, menerapkan style pada tabel
sorted_country_df.style.apply(highlight_col, axis=None)
```

```python
##membuat scatter plot
fig = px.scatter(sorted_country_df.head(25), x='country_region', y='confirmed', size='confirmed', color='country_region', hover_name='country_region', size_max=60)
```

```python
fig.show() ##menampilkan scatter plot yang telah di sort
```

```python
##Membuat tampilan visualisasi dari data terkonfirmasi
import plotly.graph_objects as go

def plot_cases_for_country(country): 
    labels = ['confirmed', 'deaths']
    colors = ['blue', 'red']
    line_size = [6, 8]
    df_list = [confirmed_df, deaths_df] 
    
    fig = go.Figure()
    
    for i, df in enumerate(df_list):
        if country =='world' or country == 'World': 
            x_data = np.array(list(df.iloc[:, 5:].columns))
            y_data = np.sum(np.asarray(df.iloc[:, 5:]), axis=0) 
        else:
            x_data = np.array(list(df.iloc[:, 5:].columns))
            y_data = np.sum(np.asarray(df[df['country']==country].iloc[:, 5:]),axis=0)
            
        fig.add_trace(go.Scatter(x=x_data, y=y_data, mode='lines+markers', 
            name = labels[i],
            line = dict(color=colors[i], width=line_size[i]), 
            connectgaps= True,
            text = "Total " + str(labels[i]) + ": " + str(y_data[-1])
            ))
    fig.show()
```

```python
plot_cases_for_country("World") ##Menampilkan data plot case
```

```python
interact(plot_cases_for_country, country='World') ##Membuat kolom
```
